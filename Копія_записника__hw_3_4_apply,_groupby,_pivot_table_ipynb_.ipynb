3.1. Створіть зводну таблицю, яка покаже, скільки покупок (інвойсів) було зроблено в кожній з філій (`Branch`) за різними категоріями товарів. Запишіть таблицю в змінну `invoices_by_category` і виведіть змінну на екран.
Ця таблиця допоможе проаналізувати, в якій філії купують найбільше товарів кожної з категорій.

```python
invoices_by_category = df.pivot_table(index='Branch', columns='Product line', values='Invoice ID', aggfunc='count')
invoices_by_category
```

 Питання можуть бути будь-якої складності, але їх має бути 3. Кожне питання оцінюється як 1 завдання. Без виконання цього завдання ДЗ не приймається. Якщо є питання щодо виконання - пишіть у чат 🙌


```python
# 1. Яка категорія товарів приносить найбільший дохід в кожному з філіалів?
category_revenue_by_branch = df.groupby(['Branch', 'Product line'])['Total'].sum().reset_index()
max_revenue_by_branch = category_revenue_by_branch.loc[category_revenue_by_branch.groupby('Branch')['Total'].idxmax()]
print(max_revenue_by_branch)
```

```python
# 2. В який день тижня спостерігається найбільший обсяг продажів у різних містах?
df['Date'] = pd.to_datetime(df['Date'])
df['Day of Week'] = df['Date'].dt.day_name()
sales_by_day_city = df.groupby(['City', 'Day of Week'])['Total'].sum().reset_index()
max_sales_by_day_city = sales_by_day_city.loc[sales_by_day_city.groupby('City')['Total'].idxmax()]
print(max_sales_by_day_city)
```

```python
# 3. Який середній обсяг продажів залежно від способу оплати?
avg_sales_by_payment = df.groupby('Payment')['Total'].mean().reset_index()
print(avg_sales_by_payment)
```
